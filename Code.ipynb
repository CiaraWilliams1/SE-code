```python
# import the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
from datetime import datetime as dt
%matplotlib inline
```

```python
# Read in files from the current working directory.
data = pd.read_csv('merged.csv', parse_dates=['DownloadDate1'])

data.head()
```

```python
data.dtypes
```

```python
# Counts by study title
AdultCYP = data.groupby(by=['AdultCYP']).count().sort_values(by=['Count'],ascending=False).reset_index()
AdultCYP
```

```python
# Counts by dataset
StudyTitle = data.groupby(by=['StudyTitle']).count()\
.sort_values(by=['Count'],ascending=False).reset_index()
StudyTitle
```

```python
# Counts by user type
UserType = data.groupby('UserType').count().sort_values(by=['Count'],ascending=False).reset_index()
UserType
```

```python
# Counts by discipline
Discipline = data.groupby('Discipline').count().sort_values(by=['Count'],ascending=False).head(10).reset_index()
Discipline
```

```python
# Counts by country
Country = data.groupby('Country').count().sort_values(by=['Count'],ascending=False).reset_index()
Country
```

```python
# Counts by institution
Institution = data.groupby('Institution').count().sort_values(by=['Count'],ascending=False).reset_index()
Institution
```

```python
# Counts by project title
ProjectTitle2 = data.groupby('ProjectTitle').count().sort_values(by=['Count'],ascending=False).reset_index()
ProjectTitle = ProjectTitle2.loc[ProjectTitle2['Count']>4]
ProjectTitle
```

```python
# Counts by project type
ProjectType = data.groupby('ProjectType').count().sort_values(by=['Count'],ascending=False).reset_index()
```

```python
# Counts by date
DownloadDate1 = data.groupby(by=['DownloadDate1']).count().sort_values(by=['Count'],ascending=False).reset_index()
```

```python
# Counts by study title
data.groupby(by=['DownloadDate1', 'Institution']).count().sort_values(by=['Count'],ascending=False).reset_index()
```

```python
# Counts by study title
data.groupby(by=['ProjectType', 'Institution']).count().sort_values(by=['Count'],ascending=False).reset_index()
```

```python
# Counts by study title
data.groupby(by=['AdultCYP', 'Institution']).count().sort_values(by=['Count'],ascending=False).head(20).reset_index()
```
